# Matplotlib

Matplotlib es una biblioteca completa para crear visualizaciones est√°ticas, animadas e interactivas en Python.

## Axes interface (object-based, explicit)

create a Figure and one or more Axes objects, then explicitly use methods on these objects to add data, configure limits, set labels etc.

### API:

subplots: create Figure and Axes

axes: add data, limits, labels etc.

Figure: for figure-level methods

### Example:

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Sample plot")
plt.show()

## pyplot interface (function-based, implicit)

consists of functions in the pyplot module. Figure and Axes are manipulated through these functions and are only implicitly present in the background.

### API:

matplotlib.pyplot

### Example:

plt.plot(x, y)
plt.title("Sample plot")
plt.show()

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# grafico de linea para ver tendencias

months = np.array(['EN', 'FE', 'MZ', 'AB', 'MY', 'JN', 'JL', 'AG', 'SE', 'OC', 'NO', 'DI'])
sales = np.array([100, 120, 140, 160, 150, 170, 200, 220, 210, 230, 240, 250])
# Figure size
plt.figure(figsize=(8, 6))

plt.plot(months, sales, marker='o', color='r', linestyle='-', linewidth=2, markersize=10)
plt.legend(['Ventas mensuales'], loc='upper left')
plt.title('Ventas mensuales')
plt.xlabel('Meses')
plt.ylabel('Ventas en miles de USD')
plt.grid(True)
plt.show()
```

```python
# Figure size
plt.figure(figsize=(8, 6))
hours = np.array([2,3,4,5,6,7,8])
exam1 = np.array([55,60,65,70,75,80,85])
plt.title('Notas segun horas de estudio')
plt.xlabel('horas de estudio')
plt.ylabel('Nota examen')
plt.grid(True)
plt.scatter(hours, exam1, color='r', s=100, marker='o')
plt.show()
```
