### Gráfico de barras:

Compara cantidades entre diferentes categorías.
Ejemplo: Comparar la venta de diferentes productos en un mes.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creación del gráfico de barras
plt.figure(figsize=(10,8))
```

```python
categories = ['Producto A', 'Producto B', 'Producto C', 'Producto D']
sales = [450, 678, 875, 325]
plt.bar(categories, sales, color='skyblue', label='Ventas mensuales')
plt.annotate('Pico de ventas', xy=(2, 875), xytext=(2.5, 800), arrowprops=dict(facecolor='black', shrink=0.025, width=1, headwidth=8))
plt.title('Ventas mensuales de productos')
plt.xlabel('Productos')
plt.ylabel('Ventas mensuales')
plt.legend()
plt.show()
```

```python
categories = ['Producto A', 'Producto B', 'Producto C', 'Producto D']
sales = [450, 678, 875, 325]
plt.barh(categories, sales, color=['skyblue','red','green', 'violet'], label='Ventas mensuales', edgecolor='black', linewidth=1.2, alpha=0.6)
plt.title('Ventas mensuales de productos')
plt.ylabel('Productos')
plt.xlabel('Ventas mensuales')
plt.grid(axis='x', linestyle='--', linewidth=0.5)
plt.show()
```

```python
products = ['Producto A', 'Producto B', 'Producto C', 'Producto D']
market_share = [25, 13, 42, 20]
plt.figure(figsize=(10,8))
plt.pie(market_share, labels=products, autopct='%1.1f%%', startangle=90, shadow=True, explode=(0, 0.05, 0, 0), colors=['skyblue','lightcoral','green', 'violet'])
plt.title('Ventas de productos')
plt.legend(products, loc='lower left')
plt.show()
```
