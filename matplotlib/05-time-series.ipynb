# SERIES DE TIEMPO

Qué son: Son datos recolectados o registrados en secuencias cronológicas, donde cada punto de datos tiene una marca temporal. Ejemplos incluyen temperaturas diarias, precios de acciones o registros de ventas.
Para qué sirven: Se utilizan para analizar y prever tendencias, patrones estacionales y ciclos a lo largo del tiempo. Son esenciales para el análisis predictivo, el monitoreo de eventos y la toma de decisiones basadas en datos históricos.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
```

```python
dates = pd.date_range('2023-01-01', periods=100)
np.random.seed(10)
values = np.random.rand(100).cumsum()
data = pd.DataFrame({'date': dates, 'value': values})
data.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(data['date'], data['value'], color='tab:green')
plt.xticks(rotation=45)
plt.title('time series data with formatted dates on x-axis')
plt.xlabel('date')
plt.ylabel('value')
plt.show()
```

```python
dates = pd.date_range('2023-01-01', periods=12, freq='ME')
values = np.random.randint(1000,5000,size=12)
sales_data = pd.DataFrame({'date': dates, 'value': values})
plt.figure(figsize=(12, 6))
plt.plot(sales_data['date'], sales_data['value'], color='skyblue', marker='o', label='Monthly Sales')
plt.title('Monthly Sales data')
plt.gca().xaxis.set_major_formatter(DateFormatter('%b %Y'))
plt.xticks(rotation=45)
plt.xlabel('date')
plt.ylabel('sales')
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```
