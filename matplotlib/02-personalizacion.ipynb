### La personalización de gráficos en Matplotlib es crucial porque:

- Claridad: Mejora la comprensión al añadir etiquetas, títulos y leyendas.
- Visualización: Utiliza colores y estilos para distinguir datos y resaltar información importante.
- Estética: Hace que los gráficos sean más atractivos y profesionales.
- Adaptabilidad: Permite ajustar gráficos para diferentes formatos y audiencias.
- Enfoque: Resalta los datos relevantes y reduce el ruido visual.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Figure size
plt.figure(figsize=(8, 6))
hours = np.array([2,3,4,5,6,7,8])
scores_student_1 = np.array([55,60,65,70,75,80,85])
scores_student_2 = np.array([58,63,68,73,78,83,88])
plt.title('Notas segun horas de estudio y estudiante')
plt.xlabel('horas de estudio')
plt.ylabel('Nota examen')
plt.grid(True)
plt.scatter(hours, scores_student_1, color='r', s=100, marker='o', label='Estudiante 1')
plt.scatter(hours, scores_student_2, color='b', s=100, marker='s', label='Estudiante 2')
plt.legend()
plt.show()
```

```python
# grafico de linea para ver tendencias
# Figure size
plt.figure(figsize=(8, 6))
hours = np.array([2,3,4,5,6,7,8])
scores_student_1 = np.array([55,60,65,70,75,80,85])
scores_student_2 = np.array([50,60,68,73,78,83,88])
plt.title('Notas segun horas de estudio y estudiante')
plt.xlabel('horas de estudio')
plt.ylabel('Nota examen')
plt.grid(True)
plt.plot(hours, scores_student_1, color='r', marker='o', label='Estudiante 1', linestyle='--', linewidth=2)
plt.plot(hours, scores_student_2, color='b',  marker='s', label='Estudiante 2', linestyle='-', linewidth=1)
plt.legend()
plt.show()
```
