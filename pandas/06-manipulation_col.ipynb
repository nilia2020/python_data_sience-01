```python
import pandas as pd
import numpy as np
file_path = 'data/online_retail.csv'
df = pd.read_csv(file_path)
df.head()
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df.head()
```

```python
df['HighPrice'] = df['TotalPrice'] > 16
df.head(10)
```

```python
df.info()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.info()
```

```python
df['DiscountedPrice'] = df['UnitPrice'] * 0.9
df.head()
```

```python
def categorize_price(price):
    if price > 5:
        return 'High'
    elif price >= 2:
        return 'Medium'
    else:
        return 'Low'
```

```python
df['PriceCategory'] = df['UnitPrice'].apply(categorize_price)
df.head(10)
```
